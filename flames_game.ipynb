```python
from tkinter import *
root=Tk()
root.geometry('350x125')
root.title('Flames Game')
root.configure(bg="blue")

#Function for removing common charaters in both the players
def remove_common_char(list1, list2):
    for i in range(len(list1)):
        for j in range(len(list2)):
            if list1[i] == list2[j]:
                c = list1[i]
                list1.remove(c)
                list2.remove(c)
                list3 = list1 + ["*"] + list2
                return [list3, True]
    list3 = list1 + ["*"] + list2
    return [list3, False]

#Function that gives relationship status
def relationship():
    #get name1 from label1
    n1 = name1.get()
    n1 = n1.upper()
    #if any spaces found, replace it with empty string
    n1.replace(" ", "")
    #create a list of charaters for name1
    n1_list = list(n1)
    #get name2 from label2
    n2 = name2.get()
    n2 = n2.upper()
    n2.replace(" ", "")
    n2_list = list(n2)
    Flag = True
    #we need to remove all common charaters by calling remove_common_char function
    while Flag:
        #store the return values in result variable
        result = remove_common_char(n1_list, n2_list)
        #get concatenated list from result
        concatenated_list = result[0]
        #get Flag value from result
        Flag = result[1]
        #find index of border mark(*)
        border_index = concatenated_list.index("*")
        #store charaters before * in n1_list
        n1_list = concatenated_list[:border_index]
        #store characters after * in n2_list
        n2_list = concatenated_list[border_index + 1:]
        #count the number of remaining characters
        count = len(n1_list) + len(n2_list)
        #flames variable
        flames = ["Friends", "Love", "Affection", "Marriage", "Enemy", "Siblings"]
        #keep looping untill last remaining item in list
        while len(flames) > 1:
            #perform list slicing
            index = (count % len(flames) - 1)
            if index > 0:
                right = flames[index + 1:]
                left = flames[:index]
                flames = right + left
            else:
                flames = flames[:len(flames) - 1]
    relationship_field.insert(10, flames[0])     
#Function to clear fields
def clear_fields():
    name1.delete(0, END)
    name2.delete(0, END)
    relationship_field.delete(0, END)
    name1.focus_set()
    
    
#Create labels for all three fields
name1 = StringVar()
name2 = StringVar()
label1 = Label(root, text = "Name1: ", bg = "yellow", font = "arial 15")
label1.grid(row=1, column=0)
label2 = Label(root, text = "Name2: ", bg = "yellow", font = "arial 15")
label2.grid(row=2, column=0)
label3 = Label(root, text = "Relationship: ", bg = "yellow", font = "arial 15")
label3.grid(row=4, column=0)
#Create entry for each label
name1 = Entry(root, textvariable = "name1", font = "italic")
name1.grid(row=1, column=1, ipadx="50")
name2 = Entry(root, textvariable = "name2", font = "italic")
name2.grid(row=2, column=1, ipadx="50")
relationship_field = Entry(root, font = "italic")
relationship_field.grid(row=4, column=1, ipadx="50")
#Create buttons
button1 = Button(root, text="Submit", bg="yellow", command=relationship)
button1.grid(row=3, column=1)
button2 = Button(root, text="Clear", bg="yellow", command=clear_fields)
button2.grid(row=5, column=1)
root.mainloop()
```
